```python
import requests
from bs4 import BeautifulSoup

url = "https://news.ycombinator.com/news"
response = requests.get(url)
```

```python
yc_web_page = response.text

#use bs4 to parse the webpage
soup = BeautifulSoup(yc_web_page,"html.parser")
soup.find(name="a")
```

```python
print(soup.prettify())
```

```python
article_text = []
article_links = []
#get upvote score for each of the articles
article_scores = [score_span.getText() for score_span in soup.find_all(name="span",class_="score")]
#convert string represented score into number format
article_scores = [int(score.split(" ")[0]) for score in article_scores]
for s in soup.select(selector="span.titleline > a"): #use > to specify direct child and direct child only of span.titleline
    print(s)
    article_text.append(s.getText())
    article_links.append(s.get("href"))
```

```python
import pandas as pd
import numpy as np
```

```python
#printing the article text and link to the article that has the highest upvote score
idmax = np.argmax(article_scores)

article_text[idmax],article_links[idmax]
```
